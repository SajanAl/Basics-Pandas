# Pivot and Pivot table 

```python
import pandas as pd
df=pd.read_csv('D:/Data/weather_cities.csv')
df
```

# Pivot help to reshape the dataframe

```python
df.pivot(index="Day",columns="City")
```

# Pivot table is used to summarize and aggregate data inside dataframe

```python
df.pivot_table(index="City",columns="Day",aggfunc="mean") #sum,count,diff,mean
```

```python
df.pivot_table(index="City",columns="Day",margins=True)
```

```python
df
```

# Change date to dateformate

```python
df['Day']=pd.to_datetime(df['Day'])
df
```

# Now apply grouper function

```python
df.pivot_table(index=pd.Grouper(freq='M',key='Day'),columns='City')
```
