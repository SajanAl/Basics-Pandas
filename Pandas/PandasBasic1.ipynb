#  Merge Dataframe

```python
sindhcity=pd.DataFrame(
    {'City':["Karachi","Larkana","Ratodero"],
     'Temprature':[30,44,46]
    }
)
sindhcity
```

```python
sindhcity2=pd.DataFrame(
    {'City':["Ratodero","Karachi","Larkana"],
     'Humidity':[6,7,8]
    }
)
sindhcity2
```

```python
weather=pd.merge(sindhcity,sindhcity2,on='City')
weather        
```

# In above we see city name same now what happen if city name not same

```python
sindhcity2=pd.DataFrame(
    {'City':["Ratodero","Karachi","Larkana","Dakhan"],
     'Humidity':[6,7,8,5]
    }
)
sindhcity2
```

```python
sindhcity=pd.DataFrame(
    {'City':["Karachi","Larkana","Ratodero","Shikarpur"],
     'Temprature':[30,44,46,40],
    }
)
sindhcity
```

```python
city=pd.merge(sindhcity,sindhcity2,on="City")
city
```

# Now apply join method outerjoin,innerjoin,rightjoin,leftjoin

```python
city=pd.merge(sindhcity,sindhcity2,on="City",how='outer')
city
```

```python
city=pd.merge(sindhcity,sindhcity2,on="City",how='right')
city
```

```python
city=pd.merge(sindhcity,sindhcity2,on="City",how='outer',indicator=True) #indicator to indicate the how to join
city
```

# Now what happen if dataframe same

```python
 sindhcity=pd.DataFrame(
    {'City':["Karachi","Larkana","Ratodero","Shikarpur"],
     'Temprature':[30,44,46,40],
     "Humidity":[5,6,7,8]
    }
)
sindhcity
```

```python
sindhcity2=pd.DataFrame(
    {'City':["Ratodero","Karachi","Larkana","Shikarpur"],
     'Temprature':[44,30,40,46],
     'Humidity':[6,7,8,5],
    }
)
sindhcity2
```

```python
df1=pd.merge(sindhcity,sindhcity2,on="City")
df1
```

```python
df1=pd.merge(sindhcity,sindhcity2,on="City",suffixes=("_left","_right"))  #suffix command handle x,y
df1
```
